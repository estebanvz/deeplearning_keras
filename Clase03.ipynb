## Importando librerias

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

## Extrayendo un datos de un Dataset

```python
dataset = keras.datasets.fashion_mnist
```

## Transformando los datos
### Partiendo los datos de entrenamiento y test

```python
((imagenes_train, etiquetas_train),(imagenes_test, etiquetas_test)) = dataset.load_data()
```

```python
print("-Dataset Entrenamiento")
print(
    "Cantidad de imágenes: {} \nCantidad de etiquetas: {} \nDimensiones de las imágenes: {}"
    .format(imagenes_train.shape[0], etiquetas_train.shape[0], imagenes_train.shape[1:])
)
print("-Dataset Test")
print(
    "Cantidad de imágenes: {} \nCantidad de etiquetas: {} \nDimensiones de las imágenes: {}"
    .format(imagenes_test.shape[0], etiquetas_test.shape[0], imagenes_test.shape[1:])
)
```

```python
print(
    "Etiquetas de entrenamiento:\n{}\nEtiquetas de test: \n{}"
    .format(list(set(etiquetas_train)),list(set(etiquetas_test)))
)
```
